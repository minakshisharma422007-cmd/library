```python
# ======= WEATHER API PROGRAM =======
# Fetching weather data from Open-Meteo API using requests module

import requests
import json
from datetime import datetime

def fetch_weather_data(latitude=52.52, longitude=13.41):
    """
    Fetches current and hourly weather data from Open-Meteo API
    
    Args:
        latitude (float): Latitude of the location (default: Berlin)
        longitude (float): Longitude of the location (default: Berlin)
    
    Returns:
        dict: JSON response containing weather data
    """
    
    # API endpoint
    url = "https://api.open-meteo.com/v1/forecast"
    
    # Query parameters
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current": "temperature_2m,wind_speed_10m",
        "hourly": "temperature_2m,relative_humidity_2m,wind_speed_10m"
    }
    
    try:
        # Make GET request
        response = requests.get(url, params=params)
        response.raise_for_status()  # Raise exception for bad status codes
        
        # Parse JSON response
        data = response.json()
        return data
        
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

def display_weather_data(data):
    """
    Displays the fetched weather data in a readable format
    
    Args:
        data (dict): Weather data from API
    """
    
    if not data:
        print("No data to display.")
        return
    
    # Display current weather
    print("\n" + "="*50)
    print("CURRENT WEATHER DATA")
    print("="*50)
    
    if "current" in data:
        current = data["current"]
        print(f"Time: {current.get('time', 'N/A')}")
        print(f"Temperature: {current.get('temperature_2m', 'N/A')}°C")
        print(f"Wind Speed: {current.get('wind_speed_10m', 'N/A')} km/h")
    
    # Display hourly weather (first 5 hours)
    print("\n" + "="*50)
    print("HOURLY WEATHER DATA (First 5 Hours)")
    print("="*50)
    
    if "hourly" in data:
        hourly = data["hourly"]
        times = hourly.get("time", [])
        temps = hourly.get("temperature_2m", [])
        humidity = hourly.get("relative_humidity_2m", [])
        wind_speeds = hourly.get("wind_speed_10m", [])
        
        # Display first 5 hours
        for i in range(min(5, len(times))):
            print(f"\nTime: {times[i]}")
            print(f"  Temperature: {temps[i]}°C")
            print(f"  Humidity: {humidity[i]}%")
            print(f"  Wind Speed: {wind_speeds[i]} km/h")

# Main execution
if __name__ == "__main__":
    # Fetch weather data for Berlin (52.52°N, 13.41°E)
    print("Fetching weather data from Open-Meteo API...")
    weather_data = fetch_weather_data()
    
    # Display the data
    if weather_data:
        display_weather_data(weather_data)
        print("\n" + "="*50)
        print("Full JSON Response:")
        print("="*50)
        print(json.dumps(weather_data, indent=2))
    else:
        print("Failed to fetch weather data.")
        print("\n" + "="*50)
```
