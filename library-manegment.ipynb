```python
from datetime import datetime, timedelta

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.borrowed_books = []  # list of (book, due_date)
    
    def borrow_book(self, book, days=14):
        if book.is_borrowed:
            print(f"Sorry, '{book.title}' is already borrowed.")
        else:
            book.is_borrowed = True
            due_date = datetime.now() + timedelta(days=days)
            self.borrowed_books.append((book, due_date))
            print(f"{self.name} borrowed '{book.title}'. Due date: {due_date.strftime('%Y-%m-%d')}")
    
    def return_book(self, book):
        for b, due_date in self.borrowed_books:
            if b == book:
                book.is_borrowed = False
                self.borrowed_books.remove((b, due_date))
                print(f"{self.name} returned '{book.title}'.")
                return
        print(f"{self.name} does not have '{book.title}'.")
    
    def list_borrowed_books(self):
        if not self.borrowed_books:
            print(f"{self.name} has not borrowed any books.")
        else:
            print(f"{self.name} has borrowed:")
            for book, due_date in self.borrowed_books:
                print(f" - {book.title} by {book.author}, due {due_date.strftime('%Y-%m-%d')}")
    
    def __str__(self):
        return f"{self.name}, Age: {self.age}"


class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False
    
    def __str__(self):
        status = "Borrowed" if self.is_borrowed else "Available"
        return f"'{self.title}' by {self.author} ({status})"


class LibrarySystem:
    def __init__(self):
        self.people = []
        self.books = []
    
    def add_person(self, name, age):
        person = Person(name, age)
        self.people.append(person)
        print(f"Added person: {person}")
        return person
    
    def add_book(self, title, author):
        book = Book(title, author)
        self.books.append(book)
        print(f"Added book: {book}")
        return book
    
    def find_person(self, name):
        for person in self.people:
            if person.name.lower() == name.lower():
                return person
        print("Person not found.")
        return None
    
    def find_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        print("Book not found.")
        return None
    
    def show_people(self):
        print("\nPeople in the system:")
        for person in self.people:
            print(person)
    
    def show_books(self):
        print("\nBooks in the system:")
        for book in self.books:
            print(book)


def main():
    system = LibrarySystem()
    
    print("Type 'help' to see available commands.")
    
    while True:
        command = input("\nEnter command: ").strip().lower()
        
        if command == "help":
            print("\n--- Library Commands ---")
            print("add person")
            print("add book")
            print("show people")
            print("show books")
            print("borrow book")
            print("return book")
            print("show borrowed")
            print("exit")
        
        elif command == "add person":
            name = input("Enter person's name: ")
            age = int(input("Enter age: "))
            system.add_person(name, age)
        
        elif command == "add book":
            title = input("Enter book title: ")
            author = input("Enter author: ")
            system.add_book(title, author)
        
        elif command == "show people":
            system.show_people()
        
        elif command == "show books":
            system.show_books()
        
        elif command == "borrow book":
            person_name = input("Enter person's name: ")
            book_title = input("Enter book title: ")
            days = input("Enter number of days to borrow (default 14): ")
            days = int(days) if days.strip() else 14
            person = system.find_person(person_name)
            book = system.find_book(book_title)
            if person and book:
                person.borrow_book(book, days)
        
        elif command == "return book":
            person_name = input("Enter person's name: ")
            book_title = input("Enter book title: ")
            person = system.find_person(person_name)
            book = system.find_book(book_title)
            if person and book:
                person.return_book(book)
        
        elif command == "show borrowed":
            person_name = input("Enter person's name: ")
            person = system.find_person(person_name)
            if person:
                person.list_borrowed_books()
        
        elif command == "exit":
            print("Goodbye!")
            break
        
        else:
            print("Unknown command. Type 'help' to see options.")


if __name__ == "__main__":
    main()
```
