```python
from datetime import datetime, timedelta

class Book:
    """Represents a book in the library."""
    def __init__(self, b_id, title, author):
        self.b_id = b_id
        self.title = title
        self.author = author
        self.available = True

class Member:
    """Represents a member of the library."""
    def __init__(self, m_id, name, age):
        self.m_id = m_id
        self.name = name
        self.age = age
        self.bwd_books = {}

    def info(self):
        print("Member:", self.name)

class Library:
    """Manages the library system."""
    def __init__(self):
        self.books = {}
        self.members = {}
        self.next_book_id = 1
        self.next_member_id = 1

    def add_book(self):
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        b_id = self.next_book_id
        self.next_book_id += 1
        self.books[b_id] = Book(b_id, title, author)
        print(f"Book '{title}' author '{author}'  added with ID {b_id}.\n")

    def add_member(self):
        name = input("Enter member name: ")
        age = input("Enter member age: ")
        m_id = self.next_member_id
        self.next_member_id += 1
        self.members[m_id] = Member(m_id, name, age)
        print(f"Member '{name}' age {age} added with ID {m_id}.\n")

    def borrow_book(self):
        title = input("Enter book title: ")
        name = input("Enter member name: ")

        book = next((b for b in self.books.values() if b.title == title), None)
        member = next((m for m in self.members.values() if m.name == name), None)

        if book and member:
            if book.available:
                book.available = False
                due_date = datetime.now() + timedelta(days=7)
                member.bwd_books[book.b_id] = due_date
                print(f"Book borrowed. Due date is {due_date.strftime('%Y-%m-%d')}\n")
            else:
                print("Book is not available.\n")
        else:
            print("Invalid book or member name!\n")

    def return_book(self):
        title = input("Enter book title: ")
        name = input("Enter member name: ")

        book = next((b for b in self.books.values() if b.title == title), None)
        member = next((m for m in self.members.values() if m.name == name), None)

        if book and member:
            if book.b_id in member.bwd_books:
                due_date = member.bwd_books.pop(book.b_id)
                book.available = True

                if datetime.now() > due_date:
                    days_late = (datetime.now() - due_date).days
                    fine = days_late * 5
                    print(f"Late return. Fine: {fine}\n")
                else:
                    print("Book returned on time.\n")
            else:
                print("This book was not borrowed by this member.\n")
        else:
            print("Invalid book or member name!\n")

    def show_books(self):
        print("\nBooks list:")
        for book in self.books.values():
            status = "available" if book.available else "borrowed"
            print(book.b_id, "-", book.title, "-", status)
        print()

    def show_members(self):
        print("\nMembers list:")
        for member in self.members.values():
            print(member.m_id, "-", member.name)
        print()
    def search_member(self):
        m_id = input("Enter member ID to search: ")
        member = next((m for m in self.members.values() if m.m_id == int(m_id)), None)
        if member:
            print(f"Member found: ID {member.m_id}, Name: {member.name}\n")
        else:
            print("Member not found!\n")

    def search_book(self):
        title = input("Enter book title to search: ")
        book = next((b for b in self.books.values() if b.title == title), None)
        if book:
            status = "available" if book.available else "borrowed"
            print(f"Book found: ID {book.b_id}, Title: {book.title}, Author: {book.author}, Status: {status}\n")
        else:
            print("Book not found!\n")

    def show_borrowers(self):
        print("\nBorrowers list:")
        for member in self.members.values():
            if member.bwd_books:
                print(member.name, "-", list(member.bwd_books.keys()))
        print()

    def show_overdue(self):
        print("\nOverdue books:")
        for member in self.members.values():
            for b_id, due_date in member.bwd_books.items():
                if datetime.now() > due_date:
                    print(member.name, "has overdue book id:", b_id)
        print()

#--------------------------------------MAIN PROGRAM------------------------------------------------------

library = Library()
print("WELCOME TO THE LIBRARY mandi system")
print("1. Add Book")
print("2. Add Member")
print("3. Borrow Book")
print("4. Return Book")
print("5. search member")
print("6. Search book")
print("7. Show Books")
print("8. Show Members")
print("9. Show Borrowers")
print("10. Show Overdue Books")
print("11. Exit")

while True:
    choice = input("Enter your choice: ").strip().lower()

    if choice == "1" or choice == "add book":
        library.add_book()
    elif choice == "2" or choice == "add member":
        library.add_member()
    elif choice == "3" or choice == "borrow book":
        library.borrow_book()
    elif choice == "4" or choice == "return book":
        library.return_book()
    elif choice == "7" or choice == "show books":
        library.show_books()
    elif choice == "8" or choice == "show members":
        library.show_members()
    elif choice == "5" or choice == "search member":
        library.search_member()
    elif choice == "6" or choice == "search book":
        library.search_book()
    elif choice == "9" or choice == "show borrowers":
        library.show_borrowers()
    elif choice == "10" or choice == "show overdue":
        library.show_overdue()
    elif choice == "11" or choice == "exit":
        print("Exiting....Thank You!")
        break
    else:
        print("Invalid Choice!")
```
