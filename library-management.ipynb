```python
from datetime import datetime,timedelta

class book:
    def __init__(self,b_id,title,author):
        self.b_id=b_id
        self.title=title
        self.auhtor=author
        self.available=True

class member:
    def __init__(self,m_id,name):
        self.m_id=m_id
        self.name=name
        self.bwd_books={}

    def info(self):
        print("member:",self.name)

class student(member):
    def __init__(self,m_id,name):
        super().__init__(m_id,name)
    def info(self):
        print("student:",self.name)

class library:
    def __init__(self):
        self.books={}
        self.members={}

    def add_book(self):
        b_id=int(input("Enter book id:"))
        title=input("Enter book title:")
        author=input("Enter author name:")
        self.books[b_id]=book(b_id,title,author)
        print(self.books)
        print("Book added.\n")

    def add_member(self):
        m_id=int(input("Enter member id:"))
        name=input("Enter member name:")
        self.members[m_id]=student(m_id,name)
        print(self.members)
        print("Member added.\n")

    def borrow_book(self):
        b_id=int(input("enter book id:"))
        m_id=int(input("enter member id:"))

        if b_id in self.books and m_id in self.members == True:
            book=self.books[b_id]
            member=self.members[m_id]
            
            if book.available==True:
                book.availbale=False
                due_date=datetime.now()+timedelta(days=7)
                member.bwd_books[b_id]=due_date
                print("book borrowed. due date is:")
            else:
                print("book is not available \n")
        else:
            print("invalid id!\n")

    def return_book(self):
        b_id=int(input("enter book id:"))
        m_id=int(input("enter member id:"))

        if b_id in self.books and m_id in self.members==True:
            book=self.books[b_id]
            member=self.members[m_id]

            if b_id in member.bwd_books:
                due_date=member.bwd_books.pop(b_id)
                book.available=True

                if datetime.now()>due_date:
                    days_late= ((datetime.now)-due_date).days
                    fine= days_late*5
                    print("late return. fine:",fine, "\n")
                else:
                    print("book returned on time")
            else:
                print("this book was not borrowed by this member \n")
            
    def show_books(self):
        print("\n books list:")
        for book in self.books.values():
            if book.available==True:
                status="available"
            else:
                status="borrowed"
            print(book.b_id,'_',book.title,'_',status)
        print()

    def show_members(self):
        print("\nmembers list:")
        for member in self.members.values():
            print(member.m_id,"-",member.name)
        print()

    def show_borrowers(self):
        print("\n borrowers list:")
        for member in self.members.values():
            if member.bwd_books:
                print(member.name,"-".list(member.bwd_books.keys()))
        print()

    def show_overdue(self):
        print("\n overdue books:")
        for member in self.members.values():
            for b_id,due_date in member.bwd_books.items():
                if datetime.now()> due_date:
                    print(member.name,"has overdue book id:",b_id)
        print()

#--------------------------------------MAIN PROGRAM------------------------------------------------------

library= library()
print ("Welcome to the Library Management System")
print("1. Add Book")
print("2. Add Member")
print("3. Borrow Book")
print("4. Return Book")
print("5. Show Books")
print("6. Show Members")
print("7. Show Borrowers")
print("8. Show Overdue Books")
print("9. Exit")

while True:
    choice=input("enter your choice:")

    if choice=="1":
        library.add_book()
    elif choice=="2":
        library.add_member()
    elif choice=="3":
        library.borrow_book()
    elif choice=="4":
        library.return_book()
    elif choice=="5":
        library.show_books()
    elif choice=="6":
        library.show_members()
    elif choice=="7":
        library.show_borrowers()
    elif choice=="8":
        library.show_overdue()
    elif choice=="9":
        print("Exiting....Thank You!")
        break
    else:
        print("Invalid Choice!")
        
```
